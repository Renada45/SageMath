## Derivative and Gradients in Multivariable Calulus

```python
# partial derivative
g(x,y) = x*y + sin(x^2) + e^(-x)
show(g(x,y))
```

```python
# Derivative respect to y
derivative(g(x,y), y) # we need to specify which variable we would like to take deriavtive 
```

```python
# Integration
integrate(g(x,y), y) # we will think it as x -> constant ; y -> variable
```

```python
plot3d(g(x,y), (x,-1,1),(y,-2,2))
```

```python
plot3d((x^3*y - y^3*x)/ 390, (x,-30,30),(y,-40,40))
```

```python
plot3d((1 - cos(sqrt(x^2 + y^2))) / sqrt(x^2 + y^2), (x,-10,10),(y,-10,10))
```
