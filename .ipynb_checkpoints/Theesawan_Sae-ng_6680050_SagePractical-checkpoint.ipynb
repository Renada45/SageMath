```python
#25 
def modified_collatz(n):
    store = []
    while (n!=1):
        if n% 2==0:
            n = n/2
            store.append(n)
        else:
            n = n*5 +1 
            store.append(n)
    for i in store:
        print(i)
        
modified_collatz(3)
```
