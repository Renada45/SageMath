```python
#26
def plot_stuff():
    a = int(input("select (a) from 1-10 :"))
    b = int(input("select (b) from 1-10 :"))
    c = int(input("select (c) from 1-10 :"))
    
    f = lambda x: a*x^2 + b*x + c
    plot_f = plot(f, (x, -10, 10))  # Plot from x = -10 to x = 10
    plot_f.show()  

plot_stuff()
```
