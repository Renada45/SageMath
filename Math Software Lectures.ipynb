# Demand Curve
$q(p) = \frac{1000}{p} -10$

```python
p = plot(1000/x-10,(x,0.1,100), ymax=600, legend_label = '$q(p) = \frac{1000}{q} -10$', axes_labels =['p (Price)', 'y = q(p) (Quantity Demanded)'])
show(p)
```
