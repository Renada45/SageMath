```python
#21
def taylor_plot():
    n =  int(input("Please enter positive integers less than 10 : "))
    f(x) = sin(x)
    taylor_series = f.taylor(x, 0, n)
    plot_f = plot(f, (x,-2,2), color='blue', legend_label='Original function')
    plot_taylor = plot(taylor_series, (x,-2,2), color='red', legend_label=f'Taylor degree of {n}')
    show(plot_f + plot_taylor)
taylor_plot()
```
